```python
import pandas as pd
import nltk
import id_nusantara
import spacy
nlp = spacy.load("id_nusantara")
```

```python
data = pd.read_csv("./dataset_sms_spam _v1.csv")
data.head()
```

```python
##tokenisasi kalimat
data['token'] = data['Teks'].apply(lambda x: nltk.word_tokenize(x))
data.head()
```

```python
data['token_kalimat'] = data['Teks'].apply(lambda x: nltk.sent_tokenize(x))
data.head()
```

```python
##lematisasi
data['lemmatize'] = data['token'].apply(lambda x: [nlp(word)[0].lemma_ for word in x])

#print lemma and token
for i in range(len(data)):
    print(data['token'][i])
    print(data['lemmatize'][i])
    print('\n')
```

```python
##POS tagging from lemma
data_PostTagging_Lemma = []

for i in range(len(data)):
    data_PostTagging_Lemma.append(nltk.pos_tag(data['lemmatize'][i]))

for i in range(len(data)):
    print(data['lemmatize'][i])
    print(data_PostTagging_Lemma[i])
    print('\n')
```

```python
#Depedency Parsing
data['DepedencyParsing'] = data['Teks'].apply(lambda x: [token.dep_ for token in nlp(x)])
data.head()


for i in range(len(data)):
    print(data['Teks'][i])
    print(data['DepedencyParsing'][i])
    print('\n')
```

```python
## case identification
## to lower case
data['lowercase'] = data['Teks'].apply(lambda x: x.lower())


for i in range (len(data)) :
    print(data['Teks'][i])
    print(data['lowercase'][i])
    print('\n')
```

```python
##morphology
data['morphology'] = data['Teks'].apply(lambda x: [token.morph for token in nlp(x)])

for i in range(len(data)):
    print(data['Teks'][i])
    print(data['morphology'][i])
    print('\n')
```

```python
## to csv
data.to_csv('data_preprocessing.csv', index=False)
```
